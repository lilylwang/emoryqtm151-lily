(a) Import and parse dates

- Import the dataset "currencies.csv"
- Parse the "date_string" column with ```pd.to_datetime()```

````python
# Write your own code
currency = pd.read_csv("data_raw/currencies.csv")

currency["date"] = pd.to_datetime(currency["date_string"], format = "%A, %b %d, %Y")


````

<font size = "5">

(b) Plot multiple time series

<font size = "3">

- Create a line plot plot which has the "gbp_to_usd" and "eur_to_usd" on the y-axis <br>
and the date on the x-axis
- Label the x-axis, the y-axis, and the title
- Add a legend and set the names to "1 Euro to USD" and "1 GPB to USD"

Hint: See Lecture 22 for how to plot multiple columns <br>
Hint: Remember to include the legend names as a list ```[...]```


````python
# Write your own code
currency = currency.sort_values("date",ascending = True)

currency[["date","gbp_to_usd","eur_to_usd"]].set_index("date").plot()
plt.title("Currency Conversion Over Time")
plt.xlabel("Time")
plt.ylabel("Currency Conversion in USD")
plt.legend(["1 GPB to USD","1 Euro to USD"])
````

<font size = "5">

(c) Compute growth rates

<font size = "3">

- Calculate a new column "growth_euro" which computes the percentage daily <br>
growth rate of "eur_to_usd".

Hint: Use ```.diff()``` and ```.shift(1)``` as in Lecture 22

````python
# Write your own code

currency = currency.sort_values("date",ascending = True)
currency["diff_euro"] = currency["eur_to_usd"].diff()


currency["lag_euro"]    = currency["eur_to_usd"].shift(1)
currency["growth_euro"] = currency["diff_euro"]* 100/currency["lag_euro"]

print(currency["growth_euro"])
````

<font size = "5">

(d) Flag subset of data and plot

<font size = "3">

- Create a boolean colum (True/False) "bool_cond" which checks whether <br>
"growth_euro" had a value greater than 2 or less than -2
- Generate a line plot with the "growth_euro" series, label the axes and the title
- Use the command ```plt.fill_between()``` as in Lecture 22 to create a shaded region <br>
for time periods where "bool_cond" is true.



````python
# Write your own code
currency["opp_growth"]  = currency["growth_euro"].between(-2,2)

currency["bool_cond"] = currency["opp_growth"] != True


# Create a line plot
plt.plot("date", "growth_euro", data = currency)
plt.xlabel("Time")
plt.ylabel("Growth Rate of the Euro")
plt.title("Currency Conversion Rate")


vec_y = currency["growth_euro"]
plt.fill_between(x= currency["date"],
                 y1 = vec_y.min(),
                 y2 = vec_y.max(),
                 where = currency["bool_cond"],
                 alpha = 0.2,color = "red")

plt.show()


````

<font size = "5">

(e) Aggregate dataset by time period

<font size = "3">

- Create an object "month_config" that uses ```pd.Grouper()```, setting the key to <br>
the column "date" and freq to "m" (month)
- Follow the instructions in Lecture 21 to compute an aggregate dataset <br>
by month which computes the standard deviation of "eur_to_usd"
- Display the dataset




````python
# Write your own code

month_config = pd.Grouper(key = "date",freq = "m")

monthly_data = currency.groupby(month_config).agg(std_monthly = ("eur_to_usd","std")).reset_index()

display(monthly_data)

plt.plot("date","std_monthly",
         data = monthly_data.sort_values("date",ascending = True))

plt.xlabel("Month")
plt.ylabel("Standard deviation")
plt.title("Evolution of the Euro to USD Exchange Rates")


````
